# Feature Engineering
#####  What is feature Engineering?
Feature engineering is the process of using domain knowledge to extract features from raw data via data mining techniques

##### Why do we need feature engineering?
It helps with:
-preparing the proper input dataset, compatible with the machine learning algorithm requirements.
-Improving the performance of machine learning models.

### Taking a look at our current dataset, we have dome redundant features, however, we can feature engineer other useful features from them. Columns such as "Year of Birth, Year of recruitment" aren't really useful to us, however, we can engineer 'Age' , "Years of Experience/working experience" and 'Employment age' ,which could be very useful to our model. Another column useful for feature engineering is:
- State of origin column

```python
#importing the required module/library
from datetime import date
import datetime
pd.Timestamp('now').year
```

```python
#creating a new column called 'Age'
train['Age'] = pd.Timestamp('now').year - train['Year_of_birth']
test['Age'] = pd.Timestamp('now').year - test['Year_of_birth']
```

```python
#creating a new column called 'Working_Years'
train['Working_Years'] = pd.Timestamp('now').year - train['Year_of_recruitment']
test['Working_Years'] = pd.Timestamp('now').year - test['Year_of_recruitment']
```

```python
#creating a new column for year at which an employee got employed - 'Employment_Age'
train['Employment_Age'] = train['Year_of_recruitment'] - train['Year_of_birth']
test['Employment_Age'] = test['Year_of_recruitment'] - test['Year_of_birth']
```

```python
#taking the fact that retirement age is 65, we feature engineer another column
def career_level(x):
    if x >= 18 and x <= 24:
        return 'Entry_level_Career'
    elif x >= 25 and x <= 34:
        return 'Mid_level_Career'
    elif x >= 35 and x <= 64:
        return 'Senior_Executives_Career'
    else:
        return 'Retirement_Age'
train['Career_Levels'] = train['Age'].apply(career_level)
career_levels =train[['Age','Career_Levels']]
```

```python
#repeat for test data as well
test['Career_Levels'] = test['Age'].apply(career_level)
career_levels =test[['Age','Career_Levels']]
```

## Evaluate Model
